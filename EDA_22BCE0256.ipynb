### **Multi-dimensional Scaling**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import euclidean_distances

# Load the dataset
url = "https://raw.githubusercontent.com/salemprakash/EDA/main/Data/school93_98.csv"
df = pd.read_csv(url)

# Limit the dataset to a smaller sample (e.g., first 100 rows or random sample)
df_sampled = df.sample(n=100, random_state=42)  # Using 100 random rows

# Step 1: Fill missing values with forward fill method
df_sampled['enrol'] = df_sampled['enrol'].ffill()
df_sampled['exppp'] = df_sampled['exppp'].ffill()
df_sampled['math4'] = df_sampled['math4'].ffill()

# Step 2: Select the numerical columns for clustering
numerical_data = df_sampled[['enrol', 'exppp', 'math4']]

# Step 3: Compute the Euclidean distance matrix
distance_matrix = euclidean_distances(numerical_data)

# Step 4: Apply Classical Multidimensional Scaling (MDS)
# Step 4.1: Double-centering the distance matrix
n = distance_matrix.shape[0]
H = np.eye(n) - np.ones((n, n)) / n  # Centering matrix
B = -0.5 * H @ (distance_matrix ** 2) @ H  # Double-centered matrix

# Step 4.2: Eigenvalue decomposition
eigvals, eigvecs = np.linalg.eigh(B)

# Step 4.3: Sort eigenvalues and eigenvectors in decreasing order
sorted_idx = np.argsort(eigvals)[::-1]
eigvals_sorted = eigvals[sorted_idx]
eigvecs_sorted = eigvecs[:, sorted_idx]

# Step 4.4: Select the top k eigenvalues and eigenvectors (2D for simplicity)
k = 2
eigvals_selected = eigvals_sorted[:k]
eigvecs_selected = eigvecs_sorted[:, :k]

# Step 4.5: Compute the coordinates in the new 2D space
X_mds = eigvecs_selected * np.sqrt(eigvals_selected)

# Step 5: Visualize the result in 2D
plt.scatter(X_mds[:, 0], X_mds[:, 1])
plt.title('Multidimensional Scaling - 22BCE0256')
plt.xlabel('MDS Component 1')
plt.ylabel('MDS Component 2')
plt.show()
```

## **Module 7 - Model Development and Evaluation**

### **Regression Analysis**

```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Load the dataset
url = "https://raw.githubusercontent.com/salemprakash/EDA/main/Data/school93_98.csv"
df = pd.read_csv(url)

# Sample the dataset to reduce size (optional)
df_sampled = df.sample(n=100, random_state=42)

# Fill missing values
df_sampled['enrol'] = df_sampled['enrol'].ffill()
df_sampled['exppp'] = df_sampled['exppp'].ffill()
df_sampled['math4'] = df_sampled['math4'].ffill()

# Define feature and target variables
X_simple = df_sampled[['enrol']]
y = df_sampled['math4']

# Split data into training and testing sets for Simple Linear Regression
X_train, X_test, y_train, y_test = train_test_split(X_simple, y, test_size=0.2, random_state=42)

# Step 1: Simple Linear Regression
simple_model = LinearRegression()
simple_model.fit(X_train, y_train)

# Predictions and evaluation
y_pred_simple = simple_model.predict(X_test)
mse_simple = mean_squared_error(y_test, y_pred_simple)
r2_simple = r2_score(y_test, y_pred_simple)

print("Simple Linear Regression")
print("Mean Squared Error:", mse_simple)
print("R-squared:", r2_simple)

# Visualization of Simple Linear Regression
plt.figure(figsize=(8, 6))
sns.scatterplot(x=X_test['enrol'], y=y_test, color='blue', label='Actual')
sns.lineplot(x=X_test['enrol'], y=y_pred_simple, color='red', label='Predicted')
plt.title("Simple Linear Regression (Enrollment vs Math4 Score) - 22BCE0256")
plt.xlabel("Enrollment")
plt.ylabel("Math4 Score")
plt.legend()
plt.show()

# Step 2: Multiple Linear Regression
X_multiple = df_sampled[['enrol', 'exppp']]
X_train, X_test, y_train, y_test = train_test_split(X_multiple, y, test_size=0.2, random_state=42)

multiple_model = LinearRegression()
multiple_model.fit(X_train, y_train)

# Predictions and evaluation
y_pred_multiple = multiple_model.predict(X_test)
mse_multiple = mean_squared_error(y_test, y_pred_multiple)
r2_multiple = r2_score(y_test, y_pred_multiple)

print("Multiple Linear Regression")
print("Mean Squared Error:", mse_multiple)
print("R-squared:", r2_multiple)

# Summary of Multiple Linear Regression using statsmodels
X_multiple_with_const = sm.add_constant(X_multiple)
model = sm.OLS(y, X_multiple_with_const).fit()
print(model.summary())
```
